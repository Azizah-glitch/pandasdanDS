```python
#S1
#running pandas dan numpy
import pandas as pd
import numpy as np

#menampilkan array 1-dimensi
ds=pd.Series([1,3,5,7,9,])
print(ds)
```

```python
#S2
data=pd.Series(['d','o','r'])
np1=np.array(['k','i','m'])
data_np=pd.Series(np1)
print(data)
print(data_np)

#mengatur indeks pada numpy array data-np di atas
data_np=pd.Series(np1,index=[9,8,7])
print(data_np)
```

```python
#S3 contoh operasi
ds1 = pd.Series([23, 87, 5, 64, 42])
ds2 = pd.Series([47, 89, 12, 63, 5])
dsa = ds1 + ds2
dsb = ds1 - ds2
dsc = ds1 * ds2
dsd = ds1 / ds2
print(dsa)
print(dsb)
print(dsc)
print(dsd)
```

```python
#S4 konver pandas module Series ke Python list
import pandas as pd
ds = pd.Series([73, 11, 58, 92, 36])
print("Pandas Series dan type")
print(ds)
print(type(ds))
print("Konver Pandas Series ke Python list")
print(ds.tolist())
print(type(ds.tolist()))
```

```python
#S5 convert a NumPy  array to a Pandas series
import numpy as np
import pandas as pd
np_array = np.array([55, 13, 92, 78, 34])
print("NumPy array:")
print(np_array)
new_np = pd.Series(np_array)
print("konversi Pandas series:")
print(new_np)
```

```python
#S6
import pandas as pd
import numpy as np

#generate angka random
num_state = np.random.RandomState(1000)  
num_series = pd.Series(num_state.normal(90, 4, 8))
print(num_series)
print(num_state)
```

# <center><div style="font-family: Times New Roman; font-size: 30px; border-radius : 10px; background-color: #a60707; color: #0f0f0fff; padding: 12px; line-height: 0.7;">Memfilter Data </div></center>

```python
#S13.2---Memilih teks
df[df['nama'].str.contains('Andi')]
df[df['nama'].str.contains('Jul')]
#Abaikan Huruf Besar/Kecil
df[df['nama'].str.contains('in')]
import re
df[df['nama'].str.contains('in', flags=re.IGNORECASE)]
#Memilih Data yang Berada di List/Dataframe Lain
feb_penjualan = [130, 220]
df[df['feb'].isin(feb_penjualan)]
```

```python
#S13---Memilih Berdasarkan Satu Kondisi
import pandas as pd
import re
#13.1---Memilih angka
#1. sama dengan
df[df['jan'] == 200]
df[df.jan.eq(200)]
#2. memilih angka tidak sama dengan
df[df['jan'] != 200]
df[df.jan.ne(200)]
df[~(df['jan'] == 200)]
#3. memilih angka lebih dari
df[df['jan'] > 200]
df[df.jan.gt(200)]
#4. Memilih lebih dari atau sama dengan
df[df['jan'] >= 200]
df[df.jan.ge(200)]
#5. Memilih kurang dari atau sama dengan
df[df['jan'] < 200]
df[df.jan.lt(200)]
#6. Memilih kurang dari atau sama dengan
df[df['jan'] <= 200]
df[df.jan.le(200)]
```

# <center><div style="font-family: Times New Roman; font-size: 30px; border-radius : 10px; background-color: #d85858ff; color: #3ea8cbff; padding: 12px; line-height: 0.7;">Metode-Metode Pandas </div></center>

```python
#P15---Metode apply untuk menjalankan fungsi pada setiap baris data series
data=[100, 150, 200, 250, 300]
d_int=pd.Series(data)
def kecilbesar(num):
    if num <120:
        return "kecil"
    elif num >=120 and num <200:
        return "sedang"
    else:
        return "besar"
d_int.apply(kecilbesar)
```
